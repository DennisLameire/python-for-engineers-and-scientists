## 2.1 - Pyplot e elementos gráficos

<br/>

#### Formato
<pre>plot(x, y, [fmt], *, data=None, **kwargs)  
retorna uma lista com o objeto Line2D </pre>

<br/>

#### sintaxe básica
```python
>>> import matplotlib.pyplot as plt
>>> x = np.linspace(0, 10, 11)
>>> y = x ** 2
>>> plt.plot(x, y, 'k--')
```

O formato base é utilizarmos um par ordenado $(x, y)$ e depois complementarmos o gráfico com outros elementos. No lugar do [fmt] podemos configurar uma cor de linha e um estilo utilizando uma string. A tabela abaixo indica alguns formatos possíveis.

<br/>

| string | descrição |
| :-- | :-- |
| '.' | ponto |
| 'o' | círculo |
| 's' | quadrado |
| '-' | linha sólida |
| '--' | linha tracejada |
| 'b' | cor azul |
| 'g' | cor verde |
| 'r' | cor vermelha |
| 'k' | cor preta |

<br/>

<pre>Podemos combinar um estilo de formatação e uma cor. 
Exemplo: '--g' = verde tracejado. </pre>

<br/>

#### Elementos adicionais

| função | descrição |
| :-- | :-- |
| plt.grid() | adiciona linhas de grade |
| plt.xlabel('titulo X') | adiciona título no eixo x |
| plt.ylabel('titulo Y') | adiciona título no eixo y |
| plt.legend() | exibe as legendas se o \*\*kwargs label é usado |
| plt.xscale(valor) | muda a escala do eixo x. Valores possíveis ```'linear', 'log', etc...  ``` |
| plt.yscale(valor) | muda a escala do eixo y. Valores possíveis ```'linear', 'log', etc...  ``` |
| plt.show() | exibe as figuras abertas |
 
<br/>

**nota:**
<pre>Use o comando abaixo para habilitar ferramentas interativas.
%matplotlib notebook</pre>

````python
import matplotlib.pyplot as plt
import numpy as np
````

````python
x = np.linspace(0, 2*np.pi, 50)
y = np.sin(x)
````

````python
plt.plot(x, y, label='xy')
plt.plot(x, x,label='xy')
plt.xlabel('x')
plt.ylabel('y')
plt.yscale('log')
plt.legend()
plt.grid()
plt.show()
````

## 2.2 - Subplots

<br/>

#### Subplots (função para quebrar a figura em subplots)
A função ```plt.subplots(nrows=1, ncols=1) ``` adiciona quadros à uma figura do matplotlib. Como resultado nós geramos uma lista com objetos onde o primeiro objeto é uma figura e o segundo é um array numpy. O array numpy é o responsável por quebrar a figura em quadros, portanto usamos as operações de unpacking para armazenar os quadros em variáveis.

<br/>

**Sintaxe:**
```python
>>> xa_arr = np.linspace(0, 2, 500)
>>> ya_arr = np.exp(xa_array)
>>>
>>> fig, (ax1, ax2) = plt.subplots(2, 2) # quebra a figura em 4 subplots
>>> ax1[1].plot(xa_arr, ya_arr) # segunda figura da primeira linha
>>> ax2[0].plot(xa_arr, ya_arr,'r-') # primeira figura da segunda linha
```

<br/>

#### Subplot (função para adicionarmos elemento em subplot)
A função ```subplot(ijn)``` é uma alternativa à ```subplots```. Neste caso nós adicionamos quadro a quadro, onde ```i``` é o número de linhas do quadro, ```j``` é o número de colunas e ```n``` é o quadro selecionado.

<br/>

**Sintaxe:**
```python
>>> xa_arr = np.linspace(0, 2, 500)
>>> ya_arr = np.exp(xa_array)
>>>
>>> plt.figure() # geramos a figura de maneira separada
>>> 
>>> plt.subplot(221) # selecionamos o primeiro quadro de uma figura com 4 quadros
>>> plt.plot(x_arr, y_arr)
>>> 
>>> plt.subplot(224) # selecionamos o quarto quadro de uma figura com 4 quadros
>>> plt.plot(x_arr, y_arr)
>>> plt.grid() # adicionamos grade para diferenciar as figuras
```




````python
import matplotlib.pyplot as plt
import numpy as np
````

````python
x = np.linspace(0, 2, 100)
ya = np.exp(xa)
yb = x ** 2
````

````python
fig, (ax1, ax2) = plt.subplots(2, 2)

ax1[0].plot(x, ya)
ax1[1].plot(x, yb, 'r')
ax2[1].plot(x, ya, 'k--')
ax2[0].grid()

plt.show()
````

````python
plt.figure()

plt.subplot(221)
plt.plot(x, ya)
plt.plot(x, yb, 'r')

plt.subplot(224)
plt.grid()
````

## 2.3 - Gráfico de barras

<br/>

#### Formato

```matplotlib.pyplot.bar(x, height, width=0.8, bottom=None, *, align='center', data=None, **kwargs)```

<br/>

#### Sintaxe

```python
>>> nomes = ['A', 'B', 'C', 'D']
>>> valores = [20,50,10,30]
>>> plt.bar(nomes, valores, color='k')
>>> plt.show()
```





````python
import matplotlib.pyplot as plt
import numpy as np
````

````python
nomes = ['A', 'B', 'C', 'D', 'E']
valores = [20, 50, 10, 2, 20]
valores2 = [10, 2, 30, 3, 10]

plt.bar(nomes, valores, color='r')
plt.bar(nomes, valores2, color='k')
plt.show()
````

````python
fig, ax = plt.subplots()

x = np.arange(5)
largura = 0.4

bar1 = ax.bar(x+largura/2, valores, color='r', width=largura, label='bar1')
bar1 = ax.bar(x-largura/2, valores2, color='k', width=largura, label='bar1')

ax.set_xticks(x)
ax.set_xticklabels(nomes)
plt.legend()
plt.show()
````

## 2.4 - Gráfico de pizzas

<br/>

#### Formato

```matplotlib.pyplot.pie(x, explode=None, labels=None, colors=None, autopct=None, pctdistance=0.6, shadow=False, labeldistance=1.1, startangle=0, radius=1, counterclock=True, wedgeprops=None, textprops=None, center=(0, 0), frame=False, rotatelabels=False, *, normalize=True, data=None) ```

<br/>

#### Sintaxe

```python
>>> nomes = ['A', 'B', 'C', 'D']
>>> valores = [20,50,10,30]
>>> plt.pie(nomes, valores)
>>> plt.show()
```


````python
import matplotlib.pyplot as plt
import numpy as np
````

````python
nomes = ['A', 'B', 'C', 'D']
valores = [20,50,10,30]

plt.pie(valores, labels=nomes)

plt.show()
````

````python
explodir = (0.5, 0, 0.1, 0.2 )


plt.pie(valores, labels=nomes, autopct='%0.2f%%', explode=explodir, shadow=True)



plt.show()
````

## 2.5 - Gráficos de superfície

<br/>

#### numpy meshgrid
O meshgrid nos auxilia na construção do gráfico de superfície. Esta função recebe um vetor de coordenadas e retorna uma matriz de de coordenadas.

#### Sintaxe
```python
>>> coord = np.meshgrid(np.array([0, 1, 2]))
```
 <br/>

#### Método add_subplot(projection='3d')

Adiciona um eixo à figura. O kwarg projection='3d' serve para passarmos a figura para 3 dimensões.
#### Sintaxe
```python
>>> ax = plt.figure().add_subplot(projection='3d')
```
 <br/>

#### Função plot_surface
Cria um plot de superfície.
#### Formato
```
plot_surface(X, Y, Z, *args, norm=None, vmin=None, vmax=None, lightsource=None, **kwargs)
```

#### Color maps (mapas de cores)
Os mapas de cores nos auxiliam na construção e leitura dos gráficos de superfície.
#### Sintaxe
```python
>>> from matplotlib import cm
>>> surf = ax.plot_surface(malha[0], malha[1], z, cmap=cm.coolwarm)
>>> # cm.coolwarm é uma escala de cores baseado em azul e vermelho
```




````python
import matplotlib.pyplot as plt
import numpy as np

from matplotlib import cm
%matplotlib notebook
````

````python
A = np.arange(3)
malha = np.meshgrid(A, A)
z = np.array([[3,2,1],
               [3,2,1],
               [3,3,3]])
````

````python
malha
````

````python
ax = plt.figure().add_subplot(projection='3d')
surf = ax.plot_surface(malha[0], malha[1], z, cmap=cm.coolwarm)
````
