## 12.1 - Pyplot and Graphical Elements

<br/>

#### Format
<pre>plot(x, y, [fmt], *, data=None, **kwargs)  
returns a list with the Line2D object </pre>

<br/>

#### Basic syntax
```python
>>> import matplotlib.pyplot as plt
>>> x = np.linspace(0, 10, 11)
>>> y = x ** 2
>>> plt.plot(x, y, 'k--')
```

The basic format is to use an ordered pair $(x, y)$ and then complement the graph with other elements. Instead of [fmt], we can configure a line color and style using a string. The table below indicates some possible formats.

<br/>

| string | description |
| :-- | :-- |
| '.' | point |
| 'o' | circle |
| 's' | square |
| '-' | solid line |
| '--' | dashed line |
| 'b' | blue color |
| 'g' | green color |
| 'r' | red color |
| 'k' | black color |

<br/>

<pre>We can combine a formatting style and a color.
Example: '--g' = dashed green. </pre>

<br/>

#### Additional Elements

| function | description |
| :-- | :-- |
| plt.grid() | adds gridlines |
| plt.xlabel('X title') | adds title on the x-axis |
| plt.ylabel('Y title') | adds title on the y-axis |
| plt.legend() | displays legends if the \*\*kwargs label is used |
| plt.xscale(value) | changes the scale of the x-axis. Possible values ```'linear', 'log', etc...  ``` |
| plt.yscale(value) | changes the scale of the y-axis. Possible values ```'linear', 'log', etc...  ``` |
| plt.show() | displays the open figures |
| plt.xlim(i, f) | sets the x-axis range from i to f |
| plt.ylim(i, f) | sets the y-axis range from i to f |

<br/>

**Note:**
<pre>Use the command below to enable interactive tools.
%matplotlib notebook</pre>

````python
import matplotlib.pyplot as plt
import numpy as np
````

````python
x = np.linspace(-1*np.pi, np.pi, 100)
sin = np.sin(x)
tan = np.tan(x)
````

````python
plt.plot(x, sin, 'r--', label='sin(x)')
plt.plot(x, tan, 'b-', label='tan(x)')
plt.xlabel('x')
plt.ylabel('y')
plt.ylim(-2, 2)
plt.xlim(-3, 3)
plt.legend()
plt.grid()
plt.show()
````

## 12.2 - Subplots

<br/>

#### Subplots (function to break the figure into subplots)
The function ```plt.subplots(nrows=1, ncols=1) ``` adds frames to a matplotlib figure. As a result, we generate a list with objects where the first object is a figure and the second is a numpy array. The numpy array is responsible for breaking the figure into frames, so we use unpacking operations to store the frames in variables.

<br/>

**Syntax:**
```python
>>> xa_arr = np.linspace(0, 2, 500)
>>> ya_arr = np.exp(xa_array)
>>>
>>> fig, (ax1, ax2) = plt.subplots(2, 2) # breaks the figure into 4 subplots
>>> ax1[1].plot(xa_arr, ya_arr) # second subplot of the first row
>>> ax2[0].plot(xa_arr, ya_arr,'r-') # first subplot of the second row
```

<br/>

#### Subplot (function to add element in subplot)
The ```subplot(ijn)``` function is an alternative to ```subplots```. In this case, we add frame by frame, where ```i``` is the number of rows of the frame, ```j``` is the number of columns, and ```n``` is the selected frame.

<br/>

**Syntax:**
```python
>>> xa_arr = np.linspace(0, 2, 500)
>>> ya_arr = np.exp(xa_array)
>>>
>>> plt.figure() # generate the figure separately
>>> 
>>> plt.subplot(221) # select the first frame of a figure with 4 frames
>>> plt.plot(x_arr, y_arr)
>>> 
>>> plt.subplot(224) # select the fourth frame of a figure with 4 frames
>>> plt.plot(x_arr, y_arr)
>>> plt.grid() # add grid to differentiate the figures
```

````python
import matplotlib.pyplot as plt
import numpy as np
````

````python
x = np.linspace(-1*np.pi, np.pi, 100)
sin = np.sin(x)
tan = np.tan(x)
exp = np.exp(x)
x2 = x ** 2
````

````python
fig, (ax1, ax2) = plt.subplots(2, 2)

ax1[0].plot(x, sin)
ax2[1].plot(x, x2, 'r--')
````

````python
plt.figure()
plt.subplot(221)
plt.grid()
plt.plot(x, sin)


plt.subplot(224)
plt.plot(x, tan)

plt.subplot(223)
plt.grid()
````

## 12.3 - Bar Plot

<br/>

#### Format

```matplotlib.pyplot.bar(x, height, width=0.8, bottom=None, *, align='center', data=None, **kwargs)```

<br/>

#### Syntax

```python
>>> names = ['A', 'B', 'C', 'D']
>>> values = [20, 50, 10, 30]
>>> plt.bar(names, values, color='k')
>>> plt.show()
```

````python
import matplotlib.pyplot as plt
import numpy as np
````

````python
names = ['A', 'B', 'C', 'D']
values = [20, 50, 10, 30]
values2 = [2*20, 50/2, 3*10, 30/3]

plt.bar(names, values, color='r')
plt.bar(names, values2, color='k')
plt.show()
````

````python
fig, ax = plt.subplots()

x = np.arange(4)
width = 0.4

bar1 = ax.bar(x-width/2, values, color='r', width=width, label='bar1')
bar2 = ax.bar(x+width/2, values2, color='k', width=width, label='bar2')

ax.set_xticks(x)
ax.set_xticklabels(names)
plt.show()
````

## 12.4 - Pie Chart

<br/>

#### Format

```matplotlib.pyplot.pie(x, explode=None, labels=None, colors=None, autopct=None, pctdistance=0.6, shadow=False, labeldistance=1.1, startangle=0, radius=1, counterclock=True, wedgeprops=None, textprops=None, center=(0, 0), frame=False, rotatelabels=False, *, normalize=True, data=None) ```

<br/>

#### Syntax

```python
>>> names = ['A', 'B', 'C', 'D']
>>> values = [20,50,10,30]
>>> plt.pie(values, labels=names)
>>> plt.show()
```

````python
import matplotlib.pyplot as plt
````

````python
names = ['A', 'B', 'C', 'D']
values = [20, 50, 10, 30]

plt.pie(values, labels=names)
plt.show()
````

````python
explode = [0.1, 0.5, 0.1, 0.1]

plt.pie(values, labels=names, autopct='%0.1f%%', shadow=True, explode=explode)
plt.show()
````

## 12.5 - Surface Plots

<br/>

#### numpy meshgrid
The meshgrid helps us in constructing the surface plot. This function takes a vector of coordinates and returns a coordinate matrix.

#### Syntax
```python
>>> coord = np.meshgrid(np.array([0, 1, 2]))
```
 <br/>

#### add_subplot(projection='3d') Method

Adds an axis to the figure. The kwarg projection='3d' is used to pass the figure to 3 dimensions.
#### Syntax
```python
>>> ax = plt.figure().add_subplot(projection='3d')
```
 <br/>

#### plot_surface Function

Creates a surface plot.
#### Format
```
plot_surface(X, Y, Z, *args, norm=None, vmin=None, vmax=None, lightsource=None, **kwargs)
```

#### Color maps
Color maps help us in constructing and reading surface plots.
#### Syntax
```python
>>> from matplotlib import cm
>>> ax = plt.figure().add_subplot(projection='3d')
>>> surf = ax.plot_surface(malha[0], malha[1], z, cmap=cm.coolwarm)
>>> # cm.coolwarm is a color scale based on blue and red
```
